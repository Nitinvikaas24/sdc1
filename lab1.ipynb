```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
import gradio as gr

# Recreate the dataset
np.random.seed(42)

n_samples = 1000

square_footage = np.random.randint(800, 3500, size=n_samples)
bedrooms = np.random.randint(1, 6, size=n_samples)
bathrooms = np.random.randint(1, 4, size=n_samples)
age_of_house = np.random.randint(1, 50, size=n_samples)

price = (square_footage * 100) + (bedrooms * 30000) + (bathrooms * 15000) - (age_of_house * 2000) + np.random.randint(-30000, 30000, size=n_samples)

df = pd.DataFrame({
    'Square Footage': square_footage,
    'Bedrooms': bedrooms,
    'Bathrooms': bathrooms,
    'Age of House': age_of_house,
    'Price (in thousands)': price
})

# Features and target
X_train = df[['Square Footage', 'Bedrooms', 'Bathrooms', 'Age of House']].values
y_train = df['Price (in thousands)'].values

# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)

# Prediction function
def predict_price(square_footage, bedrooms, bathrooms, age_of_house):
    features = np.array([[square_footage, bedrooms, bathrooms, age_of_house]])
    predicted_price = model.predict(features)[0]
    return f"Predicted House Price: ${predicted_price * 1000:.2f} (in thousands)"

# Create Gradio interface using updated syntax
inputs = [
    gr.Number(label="Square Footage"),
    gr.Number(label="Bedrooms"),
    gr.Number(label="Bathrooms"),
    gr.Number(label="Age of House")
]

outputs = gr.Textbox()

gr.Interface(fn=predict_price, inputs=inputs, outputs=outputs, title="House Price Prediction").launch()
```
